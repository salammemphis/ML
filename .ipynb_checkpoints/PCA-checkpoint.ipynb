PCA is a dimensionality reduction technique when you want to project higher dimensional data to lower dimension. It helps to reduce curse of dimenstionality. Two techniques are widely used to do PCA analysis
1. Decomposition of covariance marix 
2. SVD

The optimal number of components can be selected by looking at the eigenvalues which represents the variance explained in the new feature space. 
Total variance =summation of eigenvalues=e1+e2+e3.....+en
You can have a threshold for expected variance for the projected space and based on that you can choose number of components. 

Steps:
1. Scale the data
2. perform SVD
3. Sort the eigen vectors based on eigen values 
4. Choose number of eigen vectors based on the sum of eigen values and threshold of variance 




```python
import pandas as pd

df = pd.read_csv(
    filepath_or_buffer='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
    header=None,
    sep=',')

df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
df.dropna(how="all", inplace=True) # drops the empty line at file-end
df.drop('class', axis=1, inplace=True)
df.tail()
```

```python
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.preprocessing import StandardScaler
X=df
X_std = StandardScaler().fit_transform(X)
sklearn_pca = sklearnPCA(n_components=0.999)
Y_sklearn = sklearn_pca.fit_transform(X_std)
Y_sklearn
```

```python
#normalize and calculate mean feature wise
import numpy as np
X_std = StandardScaler().fit_transform(X)
X_mean=np.mean(X_std, axis=0)
X_mean
```

```python
#calculate covarinace matrix
X_cov=((X_std-X_mean).T.dot(X_std-X_mean))/X_std.shape[0]
X_cov
```

```python
#decompose covariance matrix
X_decpm=np.linalg.eig(X_cov)
X_decpm
```

```python
eig_vals=X_decpm[0]
eig_vecs=X_decpm[1]
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)
eig_pairs
```

```python
pc1=eig_pairs[0][1].T
pc2=eig_pairs[1][1].T
p_mat=[pc1.T,pc2.T]
p_mat=np.array(p_mat).reshape(4,2)

Projected_data=X_std.dot(p_mat)
Projected_data
```
